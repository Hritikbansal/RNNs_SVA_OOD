```python
import pickle as pkl
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
with open('RSA_LM.pkl', 'rb') as f :
    a = pkl.load(f)
```

```python
l = list(a.keys())
for k in l:
    k_ =  k.split('.pkl')[0]
    a[k_] =  a[k]
    a.pop(k)

exist={}
for k in sorted(a.keys()):
    model = k.split('_')[0]#[]
    config =k.split('_')[1]#[]
    
    if model =='LSTM':
        m='o'
    elif model =='ONLSTM':
        m='s'
    elif model =='GRU':
        m='x'
    else:
        m='^'

    if config == 'K0':
        config = "Sel"
        c = 'r'
    elif config =='K-1':
        c = 'g'
        config = "Nat"
        
    leg = model + " "+ config
    if leg in exist:
        plt.plot([a[k][0]], [a[k][1]], color=c, marker=m)
    else:
        plt.plot([a[k][0]], [a[k][1]], label = leg, color=c, marker=m)
        exist[leg]=True
    
plt.title('Representation similarity analysis in 2D', fontsize=15)
plt.xlabel('Component 1', fontsize=15)
plt.ylabel('Component 2', fontsize=15)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
